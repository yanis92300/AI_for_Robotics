```python
### Pour 1 state S nous retourne les probas de chacun de ses autres potentielles S'
def MaxActionState(probsa,V,state ):
    # 4 cases => (N,S,E,W) avec proba  dedans et on choisira le max
    result = np.zeros(4) # Nord South  East West
    for i in range(len(result)):
        for j in range(len(probsa)):            
            result[i]= result[i]+probsa[state][j][i]*V[j]
    return result
    
```

```python
def compareArray(oldV,V):
    conv=0
    for i in range(len(V)):
        if(V[i]-oldV[i]<=delta):
            conv +=1
    return conv 
        
```

```python
def  OneIteration (R):
    Pi = np.zeros(11)
    V=np.zeros(11)
    while(True):
        oldV = V.copy()
        for i in range(len(V)): # chaque case de V i = s[0]       
            V[i]=R[i]+ (Gamma*max(MaxActionState(probsa,V ,i)))
            Pi[i] = np.argmax(MaxActionState(probsa, V, i)) # NORD SUD EST WEST 
        if (compareArray(oldV,V)==11):
            break
    return (V,Pi)

```
